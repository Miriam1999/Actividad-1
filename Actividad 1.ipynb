Ejercicio 1

```python
list=[]
for i in range (1,1001,1):
    list.append(i)

divisibles=[]
no_divisibles=[]
for number in list:
    if number%7==0:
        divisibles.append(number)
    else:
        no_divisibles.append(number)
        
print(divisibles)
print(no_divisibles)
```

Ejercicio 2

```python
list=[]
for i in range (1,1001,1):
    list.append(i)

con_tres=[]
sin_tres=[]
for number in list:
    if "3" in str(number):
        con_tres.append(number)
    else:
        sin_tres.append(number)

print(con_tres)
print(sin_tres)
```

Ejericio 3

```python
text=input("Ingrese una frase: ")

list=[]
for i in text:
    list.append(i)

count=0
for character in list:
    if character==" ":
        count+=1
    else:
        continue

print("La frase tiene",count,"espacios")
```

```python
text=input("Ingrese una palabra: ")
vowels="a","e","i","o","u"

list=[]
for i in text:
    list.append(i)

for character in list:
    if character.lower() in vowels:
        list.remove(character)
    else:
        continue

print(list)
```

Ejercicio 5

```python
text=input("Ingrese una frase: ")
list=text.split(" ")

short_words=[]
for i in list:
    if len(i)<4:
        short_words.append(i)
    else:
        continue
        
print(short_words)
```
